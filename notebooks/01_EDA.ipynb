```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/home/gopaldeswal/Job-Trends-Demand-Forecasting/data/raw/jobs_raw.csv')
df.columns = df.columns.str.strip()

print(df.head())
print(df.info())
print(df.isnull().sum())


df['listed_time'] = pd.to_datetime(df['listed_time'], unit='ms', errors='coerce')


print(df['listed_time'].head())


df['listed_time'].dt.to_period('M') \
    .value_counts() \
    .sort_index() \
    .plot(kind='bar', figsize=(15, 5))

plt.title("Job Postings Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Postings")
plt.tight_layout()
plt.show()
```
